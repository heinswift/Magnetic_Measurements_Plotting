```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def f(x, y):
    return np.sin(x) ** 10 + np.cos(10 + y * x) * np.cos(x)
```

```python
x = np.linspace(0, 5, 900)
y = np.linspace(0, 5, 900)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)
```

```python
Z.shape
```

```python
plt.contourf(X, Y, Z, 20, cmap='RdGy')
plt.colorbar();
```

```python
plt.imshow(Z[::-1,:], cmap="RdGy");
```
