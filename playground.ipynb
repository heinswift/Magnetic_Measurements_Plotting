```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def f(x, y):
    return np.sin(x) ** 10 + np.cos(10 + y * x) * np.cos(x)
```

```python
x = np.linspace(0, 5, 4)
y = np.linspace(0, 5, 9)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)
```

```python
fig, ax  = plt.subplots()
plt.contourf(X, Y, Z, 200, cmap='viridis')
plt.colorbar()
fig.show()
```

```python
plt.imshow(Z[::-1,:], cmap="viridis");
```
